Demo for the dataset "Resident Evil" created by me in this notebook:

```python
import pandas as pd

games = pd.read_csv("/kaggle/input/datasets/tommasofacchin/resident-evil/games.csv")
characters = pd.read_csv("/kaggle/input/datasets/tommasofacchin/resident-evil/characters.csv")
gameAppearance = pd.read_csv("/kaggle/input/datasets/tommasofacchin/resident-evil/gameAppearance.csv")
interactions = pd.read_csv("/kaggle/input/datasets/tommasofacchin/resident-evil/interactions.csv")
scenes = pd.read_csv("/kaggle/input/datasets/tommasofacchin/resident-evil/scenes.csv")
```

```python
# We could print all the scenes from Resident Evil 2
scenes[scenes["game_id"] == "re2"]["title"]
```

## Interactions

this table contains the characters appearing in each scene. 

```python
print(f"Interactions shape: {interactions.shape}")
interactions.head()
```

```python
# We could print all Ada's scenes in Resident Evil 2.
ada_scenes = (
    interactions[(interactions["character_id"] == 3) & (interactions["game_id"] == "re2")]
    .merge(
        scenes[["game_id", "scene_id", "title"]],
        on=["game_id", "scene_id"],
        how="left"
    )
    .sort_values(["game_id", "scene_id"])
)

print(ada_scenes[["game_id", "scene_id", "title"]])
```

```python
# Or every scene where there's both Leon and Ada 
ada = interactions[interactions["character_id"] == 3][["game_id", "scene_id"]]
ada = ada.drop_duplicates()

leon = interactions[interactions["character_id"] == 1][["game_id", "scene_id"]]
leon = leon.drop_duplicates()

merged_scenes = (
    ada.merge(leon, on=["game_id", "scene_id"], how="inner")
    .merge(
        scenes[["game_id", "scene_id", "title"]],
        on=["game_id", "scene_id"],
        how="left"
    )
    .drop_duplicates()
    .sort_values(["game_id", "scene_id"])
)

print(merged_scenes[["game_id", "scene_id", "title"]])
```
